```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ETL/inmigrantes_canada.csv')
df.head()
```

```python
#Podemos hacer que nuestro indice sean los paises.
df.set_index('Pais', inplace=True)
df
```

```python
#Creando un nuevo dataFrame, que contiene en una columna los años y la cantidad de inmigrantes correspondientes.
#Creando una variable
años = list(map(str, range(1980, 2014))) #recordemos que python excluye numeros, por lo cual colocamos 2014 para que visualize hasta el 2013
años
```

```python
colombia = df.loc['Colombia', años] #localiza en el indice de "pais" a Colombia con respecto a la variable años.
colombia
#En el lado izquierdo los años son los indices y del lado derecho la cantidad de inmigrantes Colombianos en cada año.
```

```python
#Creando el dataFrame a traves de un diccionario
#Abrimos la llave, le damos un valor a la clave, en este caso la llamamos "Año", y este valor lo tomamos del indice de la variable
#que creamos anteriormente llamada "colombia" seguido de .index.totlist(), con la función index.tolist() le estamos diciendo a python que
#tome el indice de la variable y lo asigne a la clave de "Año" del diccionario.
#El mismo proceso es para nuestra clave "Inmigrantes", solo que ahora tomaremos los valores de la variable columna y ejecutamos la función
#.values.tolist()
col_dict = {'Año' : colombia.index.tolist(), 'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
#Ahora transformamos la variable creada anteriormente en un dataFrame con la siguiente linea
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

## **TRABAJANDO CON MATPLOTLIB**

```python
#Ahora vamos a visualizar un grafico, tomando en el eje "x" años y eje "Y" "inmigrantes"
plt.plot(datos_col['Año'], datos_col['Inmigrantes']) #(datos_col['X'], datos_col['Y'])
```

```python
#El grafico anterior no se entiende muy bien, por lo que podemos aplicar algunas mejoras
plt.figure(figsize=(8,4)) #podemos establecer la medida de nuestro grafico, los valores son el PULGADAS.
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) #la función plt.xticks() de la librería Matplotlib se utiliza para manipular las marcas (ticks) y etiquetas del eje x de un gráfico. Permite establecer las posiciones y etiquetas de las marcas, así como personalizar su apariencia.
plt.show()
```

```python
#COLOCANDO TITUTLOS
plt.figure(figsize=(8,4)) #podemos establecer la medida de nuestro grafico, los valores son el PULGADAS.
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) #la función plt.xticks() de la librería Matplotlib se utiliza para manipular las marcas (ticks) y etiquetas del eje x de un gráfico. Permite establecer las posiciones y etiquetas de las marcas, así como personalizar su apariencia.
plt.title('Inmigración de colombianos hacia Canada') #CREANDO EL TITULO
#Etiquetas para el eje X y Y
plt.xlabel('Años')
plt.ylabel('Inmigrantes')
plt.show() #Tiene que ser el último comando
```

```python
#Inmigrantes de BRASIL hacia Canada.
brasil = df.loc['Brasil', años] #localiza en el indice de "pais" a Colombia con respecto a la variable años.
brasil_dict = {'Año' : brasil.index.tolist(), 'Inmigrantes': brasil.values.tolist()}
datos_brasil = pd.DataFrame(brasil_dict)
datos_brasil.tail()
```

```python
#Inmigrantes de Argentina hacia Canada
arg = df.loc['Argentina', años] #localiza en el indice de "pais" a Colombia con respecto a la variable años.
arg_dict = {'Año' : arg.index.tolist(), 'Inmigrantes': arg.values.tolist()}
datos_arg = pd.DataFrame(arg_dict)
datos_arg.tail()
```

```python
plt.figure(figsize=(8,4)) #podemos establecer la medida de nuestro grafico, los valores son el PULGADAS.
plt.plot(datos_brasil['Año'], datos_brasil['Inmigrantes'])
plt.plot(datos_arg['Año'], datos_arg['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigrantes hacia Canada') #CREANDO EL TITULO
plt.legend(["Inmigrantes de Brasil","Inmigrantes de Argentina"])
#Etiquetas para el eje X y Y
plt.xlabel('Años')
plt.ylabel('Inmigrantes')
plt.show()
```

###### **¿Hay alguna tendencia o patrón común en los datos de ambos países?**
Un patrón un poco notable es que entre ambos paises, es que da la impresión que Argentina tenía mucha iniciativa en migración al inicio, empezando este pais por tener la mayor cantidad de inmigrantes hacia canada, posteriormente vemos un pico en lo años 90's, como en el 2000 y 2005, lo que el patrón de brasil es muy similar, pero varía en el año, como dando la impresión que Brasil sigue un poco la tendencia de inmigrantes a Canada con respecto a Argentina.

###### **¿Cuáles son los períodos con mayor número de inmigrantes en ambos países?**

```python
#Generando variables para almacenar los graficos
#fig almacena lienzo y ax almacena figura
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
#la siguiente linea es similar a plt.xticks, solo que esta ves es de forma automatica
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #aqui estamos diciendo que vaya en año de 5 en 5, el xaxis, estamos afectando solamente el eje x que es donde están los años.
#Añadiendo títulos y rótulos a la figura
ax.set_title('Inmigración de Colombia a Canadá\n 1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
#CREANDO SUBPLOTS EN UNA DIRECCIÓN, el objetivo es mostrar del lado izquierdo la grafica que hemos creado y del lado derecho nuestros
#datos estadisticos.
#Dentro del parentesis de subplots indicamos que queremos una fila con dos graficas, con (1,2) y agregamos nuestros figsize en pulgadas
fig, axs = plt.subplots(1, 2, figsize=(10,3))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de Colombia a Canadá\n 1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid() #para mostrar cuadriculas en nuestros graficos

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la Inmigracióm de Colombianos\nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de inmigrantes')
axs[1].grid()

plt.show()
```

```python
#Graficas para brasil, argentina.
fig, axs = plt.subplots(2, 2, figsize=(10,6))
#ajustando subplots
fig.subplots_adjust(hspace=0.5, wspace=0.3) #altura, ancho
#Creando titulo superior
fig.suptitle('Inmigración de los 4 mayores paises sudamericanos\nhacia Canada 1980 - 2013')

axs[0,0].plot(df.loc['Colombia', años])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', años])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', años])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', años])
axs[1,1].set_title('Perú')

#Estableciendo los valores de la grafica, para que todos muestren el mismo valor tanto en X como Y
ymin = 0
ymax = 7000

#iterando este mismo valor para cada grafica
for ax in axs.ravel(): #esto trae una ravel multidimensional a una ravel unidimensional
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid()
plt.show()
```

```python
#ENFOCANDO A COLOMBIA
colores = []
for pais in sudamerica_sorted.index:
  if pais == "Colombia":
    colores.append('forestgreen')
  else:
    colores.append('silver')
fig, ax = plt.subplots(figsize=(10,3))

#La siguiente linea borra el eje x
ax.set_frame_on(False) #borrando el plano cartesiano
ax.xaxis.set_visible(False) #Borrando la información del eje X, osea el valor total
#ELIMINANDO TICKS, las rayas que estan a lado de las barras.
ax.tick_params(axis='y', which='both', size=0)

#GRAFICO DE BARRAS
#agregarle h a bar (ax.barh), estamos hablando de una barra horizontal
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de Colombianos hacia Canada\n durante el periodo de 1980 - 2013', loc='left', fontsize=18)
#invertir el sentido de las x & y,
ax.set_ylabel('')
#ax.set_xlabel('Numero de inmigrantes', fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=10)

#Enumerar cada uno de los valores totales en las barras, i representa Y, J representa X.
#para i que es indice (0, 1, 2), J in enumerate, j toma el valor total de "sudamerica_sorted", en su columna "Total", mientra i representa el indice, es como contar cada barra, por ejemplo
#j toma el primer valor de la columna 'Total', es decir 75785, lo cual ese valor estará representado en el indice 0, por eso la variable i, donde el indice 0 es la barra 0. (Guayana)
for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,j,color='black', fontsize=8.5, ha='left', va='center') #ha= horizontal, va=vertical
  #EXPLICACIÓN:
  #La función enumerate() recorre una lista y te da dos cosas a la vez:
  #el índice (la posición en la lista), y
  #el valor que está en esa posición.
  #sudamerica_sorted['Total'] es una serie con los valores (los números de inmigrantes).
  #sudamerica_sorted.index contiene los nombres de los países (ya los usaste en el eje Y).
  #El gráfico usa barras horizontales, por lo tanto:
  #El eje X representa los valores (número de inmigrantes, o sea j).
  #El eje Y representa la posición de cada país en la lista, o sea i.
  #j + 20 → es la posición en X donde va el número (lo mueves 20 unidades a la derecha de la barra para que no se superponga).
  #i → es la posición en Y (la posición del país en la lista).
  #str(j) → convierte el número de inmigrantes en texto, para que se muestre como etiqueta.
  #ax.text(...) → dibuja ese texto en el gráfico.

fig.savefig('/content/sample_data/inmigrantes_sudamericanos_a_canada.png',
            transparent=False,dpi=300,bbox_inches='tight') #las siguientes funciones es para especificar como quieres exportar el png, si con fondo transparente o con fondo, en este caso conservamos el fondo
            #dpi es resolución, y bbox_inches es para especificar que margen queremos dejar, pero como no queremos dejar un margen lo dejamos como tight

plt.show()
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index,orient='h')
ax.set(title='Top 10 paises con mayor migración hacia Canada\n 1980 - 2013',
       xlabel='Numero de inmigrantes',
       ylabel='Año')
plt.show()
```

```python
#fig = lienzo, ax = figura
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Top 10 paises con mayor migración hacia Canada\n 1980 - 2013',
             loc='left', fontsize=16)
ax.set_xlabel('Numero de Immigrantes', fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def generar_grafico(palette):
#fig = lienzo, ax = figura
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False) #borrando el plano cartesiano
  ax.xaxis.set_visible(False) #Borrando la información del eje X, osea el valor total
#ELIMINANDO TICKS, las rayas que estan a lado de las barras.
  ax.tick_params(axis='y', which='both', size=0)

  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h', palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migración hacia Canada\n 1980 - 2013',
              loc='left', fontsize=16)
  ax.set_xlabel('Numero de Immigrantes', fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,j,color='black', fontsize=8.5, ha='left', va='center')
  plt.show()
```

```python
def generar_grafico(palette):
#fig = lienzo, ax = figura
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[]) #se eleminan los parametros de X, al igual de matplotlib

  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h', palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migración hacia Canada\n 1980 - 2013',
              loc='left', fontsize=16)
  ax.set_xlabel('Numero de Immigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine() #se quitan todo los bordes del grafico, linea de arriba y la derecha.
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,j,color='black', fontsize=8.5, ha='left', va='center')
  plt.show()
```
